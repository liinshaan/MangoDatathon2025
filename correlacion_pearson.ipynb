```python
import pandas as pd
df = pd.read_csv("train.csv", sep=";")
```

```python
variables_num = [
    "life_cycle_length",
    "num_stores",
    "num_sizes",
    "has_plus_sizes",
    "price",
    "year",
    "num_week_iso",
    "weekly_sales",
    "weekly_demand",
    "Production"
]

# crear Dataframe pero solo con variables numericas
df_num = df[variables_num]
```

```python
# correlacion de Pearson
matriz_corr = df_num.corr(method = "pearson")
print(matriz_corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(matriz_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlacin Pearson entre variables numricas")
plt.show()
```

Podemos eliminar:
    - weekly_demand/weekly_sales (0.95)
    - num_stores (0.71)
    - num_sizes (0.68)
